```python
import faker

# Initialize Faker generator
fake = faker.Faker()

def generate_resumes(num_resumes):
    """
    Generate synthetic resumes
    """
    resumes = [fake.text(max_nb_chars=500) for _ in range(num_resumes)]
    return resumes

def generate_job_descriptions(num_job_descriptions):
    """
    Generate synthetic job descriptions
    """
    job_descriptions = [fake.text(max_nb_chars=5000) for _ in range(num_job_descriptions)]
    return job_descriptions

# Generate synthetic data
num_resumes = 3000
num_job_descriptions = 3000

resumes = generate_resumes(num_resumes)
job_descriptions = generate_job_descriptions(num_job_descriptions)

# Save data to files
with open("resumes.txt", "w") as file:
    for resume in resumes:
        file.write(resume + "\n")

with open("job_descriptions.txt", "w") as file:
    for job_description in job_descriptions:
        file.write(job_description + "\n")
```

```python
from faker import Faker
import random

fake = Faker()

def generate_experience():
    company = fake.company()
    job_title = random.choice(["Software Engineer", "Senior Software Engineer", "Software Developer"])
    start_date = fake.date_between(start_date='-5y', end_date='-1y')
    end_date = fake.date_between(start_date='-1y', end_date='today')
    description = fake.text()

    experience = f"""
    Company: {company}
    Job Title: {job_title}
    Start Date: {start_date}
    End Date: {end_date}
    Description: {description}
    """
    return experience

def generate_resume():
    name = fake.name()
    email = fake.email()
    phone_number = fake.phone_number()
    address = fake.address()
    education = fake.text()
    skills = fake.text()

    experience = [generate_experience() for _ in range(random.randint(1, 3))]  # Generate 1 to 3 experiences

    resume = f"""
    Name: {name}
    Email: {email}
    Phone: {phone_number}
    Address: {address}

    Experience:
    {''.join(experience)}

    Education:
    {education}

    Skills:
    {skills}
    """
    return resume

# Generate a sample resume
print(generate_resume())
```
